```python
Salaries2=pd.read_sql("Select t.title,round(avg(s.salary),2) FROM Titles t Join Salaries s ON (t.emp_no = s.emp_no) group by title order by title asc", connection)
Salaries2
```

```python
Salaries3 = Salaries2.rename(columns={"title": "Title",
                                                "round": "Average Salary"})
Salaries3
```

```python
#Create a bar chart of average salary by title
# Set x axis and tick locations
x_axis = np.arange(len(Salaries3['Title']))
tick_locations = [value for value in x_axis]

# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(6,4))
plt.bar(x_axis,Salaries3["Average Salary"], color = "r", width = .5)
plt.xticks(tick_locations,Salaries3['Title'].values, rotation="vertical")
plt.title("Avg Salary by Title")
plt.xlabel("Titles")
plt.ylabel("Average Salary")
# Set x and y limits
plt.xlim(-0.75, len(x_axis)-.25)
plt.ylim(0, max(Salaries3["Average Salary"])+10)
plt.show()
```
